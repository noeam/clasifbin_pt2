```python
#df_2014_filtrado.isnull().sum()
# Definir una función para reformatear los números
def reformatear(numero):
    # Convertir el número a una cadena de texto
    texto = str(numero)
    # Si el número tiene un punto decimal, separar la parte entera y la parte decimal
    if '.' in texto:
        entera, decimal = texto.split('.')
    # Si no, asumir que la parte decimal es cero
    else:
        entera = texto
        decimal = '0'
    # Tomar los primeros dos dígitos de la parte entera como la nueva parte entera
    nueva_entera = entera[:2]
    # Tomar el resto de la parte entera y la parte decimal como la nueva parte decimal
    nueva_decimal = entera[2:] + decimal
    # Si la nueva parte decimal está vacía, poner un cero
    if nueva_decimal == '':
        nueva_decimal = '0'
    # Convertir la nueva parte entera y la nueva parte decimal a números
    nueva_entera = int(nueva_entera)
    nueva_decimal = float(nueva_decimal)
    nueva_decimal = str(nueva_decimal)
    # Dividir la nueva parte decimal por 10 elevado al número de dígitos que tiene
    nueva_decimal = nueva_decimal / (10 ** len(nueva_decimal))
    # Sumar la nueva parte entera y la nueva parte decimal para obtener el nuevo número
    nueva_decimal = float(nueva_decimal)
    nuevo_numero = nueva_entera + nueva_decimal
    # Devolver el nuevo número
    return nuevo_numero

# Aplicar la función a la columna 'numeros' y crear una nueva columna 'new_numeros'
df_2014_filtrado['IMC'] = df_2014_filtrado['imc'].apply(lambda x: float(str(x).replace('.', '')[:2] + '.' + str(x).replace('.', '')[2:]) if not pd.isna(x) else np.nan)
df_2014_filtrado
```

```python
#df_fmed_filtrado.isnull().sum()
df_fmed_filtrado['IMC'] = df_fmed_filtrado['peso_actual'] / (df_fmed_filtrado['estatura'] ** 2)
df_fmed_filtrado
```

# Comienza Transformación

# Creamos las variables categóricas para hábito ejercicio (Bueno o Malo)

```python
def exercise_habit_case(value):
    if value >= 2.5:
        return 'B'
    else:
        return 'M'
```

```python
# Leemos los valores como numericos
#df_2014_filtrado['horas_ejercicio_actual'] = pd.to_numeric(df_2014_filtrado['horas_ejercicio_actual'], errors='coerce')
#df_2014_filtrado['horas_ejercicio_1ano'] = pd.to_numeric(df_2014_filtrado['horas_ejercicio_1ano'], errors='coerce')
#df_2014_filtrado['horas_ejercicio_5ano'] = pd.to_numeric(df_2014_filtrado['horas_ejercicio_5ano'], errors='coerce')
df_2014_filtrado['ejer_act'] = df_2014_filtrado['horas_ejercicio_actual'].apply(exercise_habit_case)
df_2014_filtrado['ejer_1'] = df_2014_filtrado['horas_ejercicio_1ano'].apply(exercise_habit_case)
df_2014_filtrado['ejer_5'] = df_2014_filtrado['horas_ejercicio_5ano'].apply(exercise_habit_case)
df_2014_filtrado
```

```python
df_fmed_filtrado['ejer_act'] = df_fmed_filtrado['horas_ejercicio_actual'].apply(exercise_habit_case)
df_fmed_filtrado['ejer_1'] = df_fmed_filtrado['horas_ejercicio_1ano'].apply(exercise_habit_case)
df_fmed_filtrado['ejer_5'] = df_fmed_filtrado['horas_ejercicio_5ano'].apply(exercise_habit_case)
df_fmed_filtrado
```

# Creamos las variables categóricas para hábito cantidad de horas de sueño (Bueno o Malo)

```python
def sleep_habit_case(value):
    if (value >= 7) and (value <= 9):
        return 'B'
    else:
        return 'M'
```

```python
df_fmed_filtrado['suenio_act'] = df_fmed_filtrado['horas_suenno_calidad'].apply(sleep_habit_case)
df_fmed_filtrado['suenio_1'] = df_fmed_filtrado['suenno_diario_1ano'].apply(sleep_habit_case)
df_fmed_filtrado['suenio_5'] = df_fmed_filtrado['suenno_diario_5anos'].apply(sleep_habit_case)
df_fmed_filtrado
```

# Creamos variables de categoria Obesidad

```python
def obesity_case(value):
    if (value >= 30):
        return 1
    else:
        return 0
```

```python
df_2014_filtrado['obesidad'] = df_2014_filtrado['IMC'].apply(obesity_case)
df_2014_filtrado
```

```python
df_fmed_filtrado['obesidad'] = df_fmed_filtrado['IMC'].apply(obesity_case)
df_fmed_filtrado
```
